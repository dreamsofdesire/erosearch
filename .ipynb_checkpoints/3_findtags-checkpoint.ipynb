```python
import os
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
from rank_bm25 import *
import re
from gensim.parsing.preprocessing import STOPWORDS
from rank_bm25 import BM25Okapi
import pickle


def countToks(tokens,n=100):
    from collections import Counter
    finalToks = {}
    tokens=[x for x in tokens if len(x)>3]
    tokcount = Counter(tokens)
    for tok, count in tokcount.most_common(n):
        if len(tok) > 2:
            finalToks[tok] = count

    return finalToks
```

```python
def flattenlist(nonflatlist):
    for item in nonflatlist:
        if not isinstance(item, (str,bytes)) and isinstance(item , collections.Iterable):
            yield from flattenlist(item)
        else:
            yield item

```
